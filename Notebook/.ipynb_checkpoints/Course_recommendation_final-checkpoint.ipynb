An IT educational institute faces a challenge: the rapid evolution of IT jobs and technologies makes it difficult for students to navigate their career paths. Students often struggle to understand which job profiles align with their current skills or the technologies they wish to learn. This uncertainty can lead to confusion, indecision, and potential dropout. The institute seeks a data-driven solution to help students make informed decisions about their career paths.
Challenge:
Develop a data-driven solution that helps students understand the relationships between IT job profiles and the technologies or skills associated with them. The goal is to create a tool that can recommend suitable job profiles based on a student's existing or desired skills, reducing confusion and aiding in career planning.
1. Job Profile Clustering:
●	Cluster IT job profiles into groups based on related skills, programming languages, and frameworks.
●	Provide visualizations that help students see the connections between different job profiles and the technologies they require.
2. Recommender System:
●	Build a machine learning recommender model that suggests specific job profiles based on the skills, programming languages, or frameworks that a student has or wants to learn.
●	Tailor recommendations to individual student preferences, guiding them towards careers that match their interests and strengths.

```python
# Define a mapping dictionary for the transformation
devtype_mapping = {
    'Developer, full-stack': 'Full-Stack Developer',
    'Developer, back-end': 'Back-End Developer',
    'Developer, front-end': 'Front-End Developer',
    'Developer, mobile': 'Mobile Developer',
    'Developer, desktop or enterprise applications': 'Desktop or Enterprise Applications Developer',
    'Developer, AI': 'AI Developer',
    'Developer, game or graphics': 'Game or Graphics Developer',
    'Developer, QA or test': 'QA or Test Developer',
    'Engineer, data': 'Data Engineer',
    'Developer, embedded applications or devices': 'Embedded Applications or Devices Developer',
    'Engineer, site reliability': 'Site Reliability Engineer'}

# Function to apply the mapping
def transform_devtypes(devtype_list):
    """
    Transforms a list of DevTypes to their corresponding mapped values.

    Args:
    devtype_list (list): A list of developer types as strings.

    Returns:
    list: A list of transformed developer types.
    """
    transformed_list = []
    for devtype in devtype_list:
        # Use the mapping dictionary to replace devtypes
        transformed_devtype = devtype_mapping.get(devtype, devtype)  # Default to the original if not found
        transformed_list.append(transformed_devtype)

    return transformed_list

# Apply the transformation to the 'DevType' column
# Transform the DevTypes
transformed_devtypes = transform_devtypes(df['DevType'])
df['DevType'] = transformed_devtypes
```

### Create  a new column, a combination of all columns

```python
#create a new column containng all information on all columns
df['all_info'] = df.apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
df
```

```python
# Carry out data cleaning to remove NaN and all other unwanted
# Remove non-word characters
df['all_info'] = df['all_info'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
# Convert all to small letters
df['all_info'] = df['all_info'].apply(lambda x: x.lower())
```

```python
document = df['all_info']
document
```

### Convert a collection of documents into a matrix of TF-IDF features

```python
#Convert a collection of documents into a matrix of TF-IDF features
vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(document)
```

```python
# Print the feature matrix
print(features.toarray())

# Print the feature names (words)
print(vectorizer.get_feature_names_out())
```

### Finding best k (clusters)
Define sum of distances for each K values in range(0,10)

```python
sum_of_distances = []
K = range(0,12)
for k in K:
    kmeanModel = KMeans(n_clusters=k+1)
    kmeanModel.fit(features)
    sum_of_distances.append(kmeanModel.inertia_)
```

```python
sum_of_distances
```

### Plot sum of distances to determine appropriate k(elbow method)

```python
figure = plt.figure(figsize=(8, 5))
plt.plot(K, sum_of_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum of distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

### Clusters to use = 7

# Create a model to assign each row to a cluster

```python
k = 7
model = KMeans(n_clusters=k, init='k-means++', max_iter=100, n_init=1)
model.fit(features)

df['cluster'] = model.labels_

df.head()
```

```python
# Perform PCA to reduce the feature matrix to 2 dimensions
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(features.toarray())
n_clusters=8
# (Optional) Perform KMeans clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(features)

# Plot the clusters in a 2D scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=clusters, cmap='viridis', s=100, alpha=0.7)
plt.title('2D Scatter Plot of Clusters')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

### Visualize the different clusters

```python
#word cloud
def word_cloud(df,label):
    wc = ' '.join([text for text in df['all_info'][df['cluster'] == label]])
    wordcloud = WordCloud(width=800, height=500,
    random_state=21, max_font_size=110).generate(wc)
    fig7 = plt.figure(figsize=(10, 7))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis('off')
```

```python
# Function to predict the cluster and recommend DevTypes
def recommend_devtypes(language, webframe, tools, tech):
    # Step 2: Combine the input into 'all_info' format
    combined_info = ';'.join([language, webframe, tools, tech])

    # Step 3: Transform the new input using the pre-fitted vectorizer
    transformed_input = vectorizer.transform([combined_info])

    # Step 4: Predict the cluster using the pre-trained KMeans model
    predicted_cluster = model.predict(transformed_input)[0]

    # Step 5: Retrieve the recommended DevTypes for the predicted cluster
    recommended_devtypes = cluster_devtype_map['DevType'][predicted_cluster]

    return predicted_cluster, recommended_devtypes
```

```python
#Applying the function
language_input = 'Bash/Shell'
webframe_input = ' '
tools_input = 'CMake;Docker;GNU GCC;LLVM;Clang;Ninja;npm'
tech = 'NumPy;Pandas'

cluster, devtypes = recommend_devtypes(language_input, webframe_input, tools_input, tech)

print(f"The student belongs to cluster {cluster}.")
print("Recommended DevTypes:", devtypes)
```
